```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd
```

```python
search_query = "https://www.youtube.com/watch?v=-Y9VtoPvtuM&list=RDaPzqSS71Oys&index=14"
driver = webdriver.Chrome()
title = []
channel = []
view =[]
year = []
driver.get(search_query)
time.sleep(5)
link_locator = driver.find_elements(By.ID , "dismissible")
for each in link_locator:
    temp_title = each.find_element(By.ID, "video-title")
    temp_channel = each.find_element(By.ID, "text-container")
    temp_view = each.find_element(By.ID , "metadata-line")
    text_view = temp_view.text
    if "\n" not in text_view: 
        view.append("0")
        year.append(text_view)
    else:
        x = text_view.split("\n")
        view.append(x[0])
        year.append(x[1])
    
    title.append(temp_title.text)
    channel.append(temp_channel.text)
   

```

```python
df = pd.DataFrame({'title':title,'channel':channel,'view':view,'year':year})
df
```
